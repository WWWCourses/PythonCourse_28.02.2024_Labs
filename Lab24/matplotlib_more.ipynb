```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create some data to be used in next examples:
x = np.linspace(1,100,10)
x
```

```python
# Create 2 axes, stacked vertically:
fig, [ax1,ax2] = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10,6), dpi=200)
# plot lines
ax1.plot(x,x**2, color="red")
ax2.plot(x,x**3, color="green")

# set axes titles:
ax1.set_title('x**2')
ax2.set_title('x**3')

plt.tight_layout()

plt.savefig('figure_1280720.jpg')
```

## 3D Ploting

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Define data
x = [1, 2, 3]
y = [4, 5, 6]
z = [7, 8, 9]

# Plot the line
ax.scatter(x, y, z)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Basic 3D Line Plot')
plt.tight_layout()
```

```python
x = np.linspace(-2,2,5)
y = np.linspace(-2,2,5)
print(x)
print(y)
```

```python
xx,yy = np.meshgrid(x,y)

zz = xx+yy

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surface = ax.plot_surface(xx,yy,zz,  cmap="Accent")

# Add a color bar which maps values to colors.
fig.colorbar(surface, shrink=0.5)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the Rosenbrock function
def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

# Generate data for the surface plot
x = np.linspace(-2, 2, 100)
y = np.linspace(-1, 3, 100)

X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)

# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Surface Plot')

fig.colorbar(surface, shrink=0.5)
```
