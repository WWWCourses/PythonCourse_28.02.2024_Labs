```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,11)
```

```python
# TODO: why y tick are multiplicated
# create 2 axes, sacked vertically:
# fig = plt.figure()
fig, [ax1, ax2] = plt.subplots(nrows=2, ncols=1, sharex=False, figsize = (10,6), dpi=100)
ax1= fig.add_subplot(211)
ax2 = fig.add_subplot(212)
# plot lines
ax1.plot(x,x**2, color="red")
ax2.plot(x,x**3, color="green")
# set axes titles
ax1.set_title("x**2")
ax1.set_title("x**3")
# plt.tight_layout()
plt.savefig('figure_1280x720.jpg')
```
