```python
# to use %matplotlib widget => pip install ipympl
# %matplotlib widget

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 20, 10, dtype=np.int16)
y = x**2

print(f'x: {x}')
print(f'y: {y}')
```

```python
fig, [ax1,ax2] = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax1.plot(x, y)
ax2.plot(x, x**3)

ax1.set_title('y=x**2')
ax2.set_title('y=x**3')

ax1.set_xlabel("X",fontsize=14, color="darkred", loc="right")
ax1.set_ylabel("Y", rotation=60,fontsize=14, color="darkred", loc="top")

ax1.get_xticklabels()[3].set_color("red")
```

```python
fig, ax = plt.subplots()
print(x)
ax.plot(x, x**2, label="y=x**2", color="red")
ax.plot(x, x**3, label="y=x**3", color="green")
ax.legend(loc = 'upper left', fontsize = 'large', shadow=True, framealpha = 0)

```

```python
#  Plot line using State-based interface (pyplot api):
x = np.linspace(0,10,100)

# plot on the 1st axes:
plt.subplot(2, 1, 1)
plt.title('sin')
plt.plot(x, np.sin(x), color="#F00")

# plot on the 2nd axes:
plt.subplot(2, 1, 2)
plt.title('cos')
plt.plot(x, np.cos(x), color="#0F0")
plt.show()
```
