```python
from matplotlib import pyplot as plt
from matplotlib import colormaps
import numpy as np
```

```python
# Sample Data (tempreture per day of a week)
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [22, 23, 24, 22, 20, 19, 21]

# Creating a line plot
plt.plot(days, temperatures, marker='o', linestyle='-', color='b')
plt.title("Weekly Temperature Variation")
plt.xlabel("Days of the Week")
plt.ylabel("Temperature (°C)")
plt.grid(True)
```

```python
# Sample Data (students' test scores)
math_scores = [85, 70, 92, 78, 88, 95, 60, 75, 85, 70]
science_scores = [75, 65, 88, 72, 80, 92, 55, 70, 78, 68]

# Creating a scatter plot
plt.scatter(math_scores, science_scores, c='blue', marker='o', label='Students')
plt.title("Math vs. Science Test Scores")
plt.xlabel("Math Scores")
plt.ylabel("Science Scores")
plt.legend()
plt.grid(True)
```

```python
# histogram:
# [2-6]
# [2-3], [3-4], [4-5], [5-6],[6,]
#
scores = np.array([2,2,2,2,2,3,3,4,5,5,5,6,6,6,6,6,6],dtype="int")
print(f'min: {min(scores)}')
print(f'max: {max(scores)}')
print(f'len: {len(scores)}')
print(np.histogram(scores,bins=4))

plt.hist(scores, bins=5, stacked=True)
```

```python
scores = np.array([2,2,2,2,2,3,3,4,5,5,5,6,6,6,6,6,6],dtype="int")
plt.boxplot(scores)

print(f'median: {np.median(scores)}')
print(f'quartiles: {np.quantile(scores, [0.25, 0.5, 0.75 ,1])}')
```

```python
# Box plot
# Sample data (total sales of a product)
products = ['toys', 'food', 'drinks']
sales = [100, 50, 200]

# Create the bar plot
plt.bar(products, sales, color = ['red', 'green', 'blue'], edgecolor='black')

# Add labels and title
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Product Sales")
```

```python
# pie chart
# Sample Data
budget_categories = ['Housing', 'Food', 'Transportation', 'Entertainment', 'Savings']
budget_allocation = [5, 5, 10, 2, 15]

# Creating a pie chart
plt.pie(budget_allocation, labels=budget_categories, autopct='%1.1f%%')
plt.title("Monthly Budget Allocation")
```
